```python
import networkx as nx
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/modified/concat/concat_minimal.csv')
```

```python
G = nx.from_pandas_edgelist(df, source='sw', target='ew', 
                            edge_attr='어휘', create_using=nx.DiGraph())
```

```python
nx.degree_centrality(G)
```

```python
nx.shortest_path(G, source='가', target='팅')
```

```python
nx.draw(G, with_labels=True, font_family='Pretendard Variable')
```
